```
import numpy as np
from numpy.linalg import inv
x1 = np.array([1.6, 22, 33, 15.5, 43, 40])
x2 = np.array([851, 1058, 1357, 816, 1201, 1115])
y = np.array([193, 172, 113, 230, 91, 125])
k, N = 2, len(y)
X = np.column_stack((np.ones(N), x1, x2)) # Using (10)
# Finding Coefficient Matrix
A = np.matmul(X.T, X)
g = np.matmul(X.T, y)
b = np.matmul(inv(A), g) # Using (12)
# Finding Rsq
y_reg = b[0] + b[1]*x1 + b[2]*x2
SSE = sum((y-y_reg)**2)
SST = sum((y-np.mean(y))**2)
Rsq = 1 - SSE/SST
print("Matrix, X =\n", X)
print("\nRegression Coefficients --> ")
print(" b[2] =", b[2])
print(" b[1] =", b[1])
print(" b[0] =", b[0])
print("\nEstimated Relation:\n y = {} + {} x1 + {} x2".format(b[0], b[1], b[2
]))
print("\nR-squared =", Rsq)
```
